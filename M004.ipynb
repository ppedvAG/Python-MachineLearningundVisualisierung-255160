### N.A.

```python
data == "N.A."
```

```python
(data == "N.A.").value_counts()  # Mit value_counts() Spalten mit Fehlern finden
```

Allgemein ist es schwierig, alle fehlerhaften Werte im ganzen DataFrame gleichzeitig auszutauschen

Lösung: Eigene Funktion, die einen Spaltennamen empfängt

```python
data["FertRate"] == "N.A."
```

```python
zeilen = data[data["FertRate"] == "N.A."]
```

```python
zeilen["FertRate"] = 0  # Warnung
```

```python
zeilen.loc[:, "FertRate"] = 0  # Neue Werte eintragen
```

```python
zeilen
```

```python
index = zeilen.index
```

```python
data.loc[index, "FertRate"] = 0
```

```python
data
```

#### Eigene Funktion

```python
def fixNA(colName, newValue):
    x = data[colName] == "N.A."  # Alle fehlerhaften Zeilen finden
    i = x[x == True].index  # Die Indizes der fehlerhaften Zeilen entnehmen
    data.loc[i, colName] = newValue  # Mit loc und dem Index die originalen Zeilen mit newValue beschreiben
```

```python
for col in data.columns:
    fixNA(col, 0)
```

```python
data
```

### Prozentzeichen

Für Prozentzeichen kann str.replace verwendet werden

Diese Funktion kommt direkt von der Spalte

```python
data["WorldShare"].replace(" %", "", regex=True)  # Hier muss regex=True aktiviert sein
```

```python
data["WorldShare"] = data["WorldShare"].replace(" %", "", regex=True)
```

```python
data
```

#### Funktion für Prozent definieren

```python
def fixPct(colName):
    data[colName] = data[colName].replace(" %", "", regex=True)
```

```python
for col in data.columns:
    fixPct(col)
```

```python
data
```

#### Funktion für Typen definieren

```python
def fixType(colName, newType):
    data[colName] = data[colName].astype(newType)
```
