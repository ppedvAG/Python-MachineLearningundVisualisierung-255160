### Globale vs. Objektorientierte Graphen

Globaler Modus: plt.plot(), plt.legend(), plt.xlabel(), plt.ylabel(), plt.show(), ...

Objektorientierter Modus: plt.figure() oder plt.subplots()

```python
figure, ((a1_1, a1_2), (a2_1, a2_2)) = plt.subplots(2, 2)
a1_1.plot(data["Open"])
a1_1.set_title("Open")

a1_2.plot(data["Volume"])
a1_2.set_title("Volume")
```

```python
plt.plot(data["Open"])
plt.xlabel("Open")
plt.xticks(rotation=90)
plt.show()

plt.plot(data["Volume"])
```

```python
plt.bar(data.index, data["Open"])

plt.savefig("Test.png")
```
